# Ray Theory, Travel-Time Tomography, Earthquake Location

# 1. Ray Tracing and the WKB Approximation 

## 1.1 Scalar wave equation

$$
\nabla^2 u - \frac{1}{c^2(\mathbf{x})}\frac{\partial^2 u}{\partial t^2}=0.
$$
We assume a plane wave  $u=A(\mathbf{x})e^{i\omega(T(\mathbf{x})-t)}$.
Plugging in:
$$
\nabla^2(Ae^{i\omega(T-t)}) + \frac{\omega^2}{c^2}Ae^{i\omega(T-t)}=0.
$$

Expanding:
$$
\nabla^2A + 2i\omega \left[ \nabla A\cdot\nabla T + A\nabla^2T \right] - \omega^2 \left[ A|\nabla T|^2 + \frac{1}{c^2}A \right] = 0.
$$

## 1.2 Separate by powers of frequency

### (a) High frequency Approximation

Order ($\omega^2$): Eikonal equation

$$
|\nabla T|^2 = \frac{1}{c^2(\mathbf{x})}.
$$

This defines travel time $(T(\mathbf{x}))$ and $(\mathbf{p}=\nabla T)$ is the slowness vector.

### (b) Order ($\omega$): Transport equation

The transport equations will constrain the amplitude decay as a function of distance.

$$
2\nabla A\cdot\nabla T + A\nabla^2 T = 0.
$$
Along a ray path where $(\frac{d\mathbf{x}}{ds} = \frac{\nabla T}{|\nabla T|})$:$
$$
\frac{dA}{ds} = -\frac{A}{2}\nabla^2 T , \frac{1}{|\nabla T|}.
$$
Integrating gives $(A^2J = \text{const})$, with $(J)$ the geometric spreading (Jacobian of nearby rays).


## 1.3 Ray equations as a Hamiltonian system

The Hamiltonian formulation provides a powerful framework for understanding seismic wave propagation and ray tracing. Here's a detailed breakdown:

### What is a Hamiltonian?

The Hamiltonian $H$ is a function that describes the total energy of a system in terms of positions and momenta. In seismic wave propagation:

$$
H(\mathbf{x}, \mathbf{p}) = \frac{1}{2}\sum_i p_i^2 - \frac{1}{2}c^2(\mathbf{x})|\mathbf{p}|^2
$$

where:
- $\mathbf{x}$ are the spatial coordinates
- $\mathbf{p}$ are the components of slowness vector (momentum)
- $c(\mathbf{x})$ is the velocity field

### Hamilton's Equations

Define Hamiltonian $H(\mathbf{x},\mathbf{p})=\frac12(c^2|\mathbf{p}|^2-1)$.
Then the rate of change in position and wavenumber is:
$$
\frac{d\mathbf{x}}{dt} = \frac{\partial H}{\partial \mathbf{p}} = c^2\mathbf{p}, \quad
\frac{d\mathbf{p}}{dt} = -\frac{\partial H}{\partial \mathbf{x}} = -(\nabla c)c|\mathbf{p}|^2.
$$


The Hamiltonian formulation is equivalent to the Eikonal equation: The slowness vector $\mathbf{p} = \nabla T$ and the Hamiltonian constraint $H=0$ for physical ray paths (e.g., $\mathbf{x}$ is along ray paths).

With $(|\mathbf{p}|=1/c)$, we get the **ray curvature** equation:
$$
\frac{d\hat{\mathbf{n}}}{ds} = \nabla\ln c - (\hat{\mathbf{n}}\cdot\nabla\ln c)\hat{\mathbf{n}}.
$$


The ray paths are determined by Hamilton's canonical equations:

$$
\begin{align}
\dot{\mathbf{x}} &= \frac{\partial H}{\partial \mathbf{p}} \\
\dot{\mathbf{p}} &= -\frac{\partial H}{\partial \mathbf{x}}
\end{align}
$$

These equations give us:
1. The rate of change of position ($\dot{\mathbf{x}}$) is the group velocity
2. The rate of change of slowness ($\dot{\mathbf{p}}$) depends on velocity gradients


## 1.4 Analytical examples

### (a) Homogeneous medium 

Assume $c=c_0$.

Eikonal: $(|\nabla T|=1/c_0)$.
Solution: $(T(\mathbf{x}) = |\mathbf{x}-\mathbf{x}_s|/c_0)$.
Rays: straight lines.

Amplitude:
$$
A \propto
\begin{cases}
1/\sqrt{r} & 2\text{-D}\
1/r & 3\text{-D.}
\end{cases}
$$

### (b) Linear velocity gradient 

Assume $(c(z)=c_0+\gamma z)$.

Horizontal slowness $p$ conserved: $(p=\sin\theta/c(z))$.
Then $ \cos\theta(z)=\sqrt{1-p^2c^2(z)}$ and we integrate:

$$
x(z) = \int \frac{pc(z)}{\sqrt{1-p^2c^2(z)}}dz,
\qquad
T(z) = \int \frac{dz}{c(z)\sqrt{1-p^2c^2(z)}}.
$$

Substitute $(c(z)=c_0+\gamma z$)$ ⇒ logarithmic form:
$$
x(z)=\frac{1}{\gamma p}\sqrt{1-p^2c^2(z)}-\frac{1}{\gamma p}\sqrt{1-p^2c_0^2}.
$$
Travel time follows similarly (elliptic or log term). Rays are circular arcs.


### (c) Layered medium 

Assume $c=c(z)$, with piece-wise constant velocities.

Snell’s law: $p=\sin\theta/c(z)=\text{const}$.

Between two interfaces $(z_1,z_2)$:
$$
\Delta x = \int_{z_1}^{z_2} \frac{pc(z)}{\sqrt{1-p^2c^2(z)}},dz,
\qquad
\Delta T = \int_{z_1}^{z_2} \frac{1}{c(z)\sqrt{1-p^2c^2(z)}},dz.
$$

For a source and receiver both at the surface $(z=0)$, with horizontal offset $X$, we need to:

1. **Determine maximum depth**: For a given source-receiver distance $X$, the ray parameter $p$ must satisfy:
   $$
   p \leq \frac{1}{v_{\text{max}}}
   $$
   where $v_{\text{max}}$ is the maximum velocity encountered along the ray path.

2. **Ray path equations**: For each layer $i$ with velocity $c_i$ and thickness $h_i$:
   $$
   \Delta x_i = \frac{2ph_i}{\sqrt{1-p^2c_i^2}}
   $$
   $$
   \Delta T_i = \frac{2h_i}{c_i^2\sqrt{1-p^2c_i^2}}
   $$
   where the factor of 2 accounts for down-going and up-going paths.

3. **Finding the correct $p$**: Solve for $p$ that satisfies:
   $$
   X = \sum_{i=1}^n \Delta x_i(p)
   $$
   This is typically done numerically since the equation is nonlinear.

4. **Turning depth**: For a given $p$, the turning depth occurs in the layer where:
   $$
   c(z_{\text{turn}}) = \frac{1}{p}
   $$

### Example for two-layer case:
For a model with:
- Layer 1: $0 \leq z \leq h_1$, velocity $c_1$
- Layer 2: $h_1 \leq z \leq h_2$, velocity $c_2 > c_1$

The total horizontal distance $X$ is:
$$
X = \begin{cases}
\frac{2ph_1}{\sqrt{1-p^2c_1^2}} & \text{if ray turns in layer 1} \\
\frac{2ph_1}{\sqrt{1-p^2c_1^2}} + \frac{2ph_2}{\sqrt{1-p^2c_2^2}} & \text{if ray turns in layer 2}
\end{cases}
$$

The corresponding travel time $T$ is:
$$
T = \begin{cases}
\frac{2h_1}{c_1^2\sqrt{1-p^2c_1^2}} & \text{if ray turns in layer 1} \\
\frac{2h_1}{c_1^2\sqrt{1-p^2c_1^2}} + \frac{2h_2}{c_2^2\sqrt{1-p^2c_2^2}} & \text{if ray turns in layer 2}
\end{cases}
$$

```python
import numpy as np
from scipy.optimize import root_scalar
import matplotlib.pyplot as plt

def compute_X_T(p, layers):
    """
    Compute total horizontal distance X and travel time T for a given ray parameter
    
    Parameters:
    -----------
    p : float
        Ray parameter (horizontal slowness)
    layers : list of tuples
        Each tuple contains (thickness, velocity) for each layer
    
    Returns:
    --------
    X : float
        Total horizontal distance
    T : float
        Total travel time
    """
    X = 0
    T = 0
    
    for h, c in layers:
        # Skip layer if ray parameter exceeds critical value
        if p > 1/c:
            return np.inf, np.inf
        
        # Add contribution from this layer
        X += 2 * p * h / np.sqrt(1 - p**2 * c**2)
        T += 2 * h / (c**2 * np.sqrt(1 - p**2 * c**2))
    
    return X, T

# Define a two-layer model
layers = [
    (2.0, 3.0),  # 2 km thick, 3 km/s
    (3.0, 5.0)   # 3 km thick, 5 km/s
]

# Range of source-receiver distances to consider
X_range = np.linspace(0.1, 20, 50)
T_values = []

for X_target in X_range:
    # Define objective function for root finding
    def objective(p):
        X, _ = compute_X_T(p, layers)
        return X - X_target
    
    # Find ray parameter that gives desired X
    try:
        # Bracket the solution between minimum and maximum possible p
        p_max = 1/min(v for h, v in layers)
        result = root_scalar(objective, bracket=[1e-6, p_max])
        p = result.root
        
        # Compute travel time for this ray parameter
        _, T = compute_X_T(p, layers)
        T_values.append(T)
    except:
        T_values.append(np.nan)

# Plot travel time curve
plt.figure(figsize=(10, 6))
plt.plot(X_range, T_values, 'b-', label='Travel Time Curve')
plt.xlabel('Source-Receiver Distance (km)')
plt.ylabel('Travel Time (s)')
plt.title('Travel Time Curve for Two-Layer Model')
plt.grid(True)
plt.legend()
plt.show()
```

#### Understanding the Implementation

The code above demonstrates:

1. **Ray Parameter Search**: For each source-receiver distance $X$, we find the ray parameter $p$ that satisfies the total distance equation using root-finding.

2. **Layer Integration**: For each layer, we compute:
   - The horizontal distance contribution $\Delta x_i$
   - The travel time contribution $\Delta T_i$
   
3. **Critical Angles**: The code handles critical ray parameters where $p = 1/c_i$ for any layer $i$. Beyond these values, the equations become complex.

4. **Travel Time Curve Features**:
   - Near-offset (small $X$): Nearly vertical rays, $T \approx 2z/c_1$
   - Far-offset: Rays turning in deeper layers, showing characteristic curve shape
   - Discontinuities in $dT/dX$ indicate critical angles

This implementation can be extended to:
- Add more layers
- Include velocity gradients within layers
- Calculate amplitude effects
- Handle diving waves and head waves

## 1.5 Amplitude from geometrical spreading

Let $J(s)$ be the cross-sectional area of a ray tube.

Energy conservation constrains that  $A^2 v_g J = \text{const}$, with $v_g=c$.
$$
A(s) = A_0\left(\frac{J_0c_0}{J(s)c(s)}\right)^{1/2}.
$$

In 2-D, $J\propto \partial x_\perp/\partial\theta_0$, often computed numerically via paraxial rays.


# 2. Numerical Ray Tracing Methods

There are three primary approaches to solving ray equations numerically:

## 2.1 Shooting Methods

The shooting method works by:
1. Starting from a source point
2. Choosing an initial ray direction (takeoff angle)
3. Integrating the ray equations forward
4. Iteratively adjusting the initial angle until the ray hits the target

**Advantages:**
- Simple to implement
- Works well for simple velocity models
- Natural for computing multiple arrivals

**Disadvantages:**
- May not find all possible paths
- Can be unstable for complex velocity structures
- Requires iterative solving for two-point problems

## 2.2 Bending Methods

The bending method works by:
1. Starting with an initial ray path guess
2. Computing path integral of travel time
3. Using variational principles to minimize travel time
4. Iteratively perturbing the path until Fermat's principle is satisfied

**Advantages:**
- More stable for complex velocity models
- Directly solves two-point problem
- Often faster convergence

**Disadvantages:**
- May converge to local minimum
- More complex to implement
- Harder to find multiple arrivals

## 2.3 Grid-Based Eikonal Solvers

Made popular with dense seismic arrays, this method directly solves the eikonal equation on a grid:
1. Discretize the eikonal equation $|\nabla T|^2 = 1/c^2(\mathbf{x})$ on a regular grid
2. Use finite-difference schemes (e.g., Fast Marching Method or Fast Sweeping Method)
3. Compute travel times to all points in the medium
4. Extract ray paths by following $\nabla T$ backward from receivers

**Advantages:**
- Computes travel times to all points at once
- Very stable and robust
- Handles complex velocity structures well
- No ray tracing required for travel times

**Disadvantages:**
- May miss some multipathing effects
- Resolution limited by grid spacing
- Can be computationally intensive for large 3D grids
- Amplitude calculations require additional steps


# 3. 2-D Ray Tracer with shooting-ray method

Below is a minimal script that integrates ray equations for a 2-D isotropic medium $c(x,z)$ using the shooting method.

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# --- velocity model and its gradient ---
def velocity_2D(x, z, c0=3.0):
    """
    2D velocity model c(x,z) [km/s] with sinusoidal lateral variations
    
    Parameters:
    -----------
    x, z : float
        Position coordinates [km]
    c0 : float
        Reference velocity at origin [km/s]
    """
    return c0 * (1.0 + 0.1*np.sin(0.5*x) + 0.2*z)

def grad_c(x, z, h=0.1, x_bounds=(-10, 10), z_bounds=(0, 10)):
    """
    Compute gradient of velocity [dc/dx, dc/dz] using adaptive finite differences.
    Uses 4th order central differences when possible, falls back to 2nd order
    central, forward, or backward differences near boundaries.
    
    Parameters:
    -----------
    x, z : float
        Position coordinates [km]
    h : float
        Grid spacing for finite difference [km]
    x_bounds : tuple
        (min_x, max_x) domain boundaries
    z_bounds : tuple
        (min_z, max_z) domain boundaries
        
    Returns:
    --------
    array : [dc/dx, dc/dz]
    """
    # Helper function for different finite difference schemes
    def central_4th(f, x, h, coord='x'):
        """4th order central difference"""
        if coord == 'x':
            return (-f(x+2*h, z) + 8*f(x+h, z) - 8*f(x-h, z) + f(x-2*h, z)) / (12*h)
        else:  # coord == 'z'
            return (-f(x, z+2*h) + 8*f(x, z+h) - 8*f(x, z-h) + f(x, z-2*h)) / (12*h)
    
    def central_2nd(f, x, h, coord='x'):
        """2nd order central difference"""
        if coord == 'x':
            return (f(x+h, z) - f(x-h, z)) / (2*h)
        else:  # coord == 'z'
            return (f(x, z+h) - f(x, z-h)) / (2*h)
    
    def forward_2nd(f, x, h, coord='x'):
        """2nd order forward difference"""
        if coord == 'x':
            return (-f(x+2*h, z) + 4*f(x+h, z) - 3*f(x, z)) / (2*h)
        else:  # coord == 'z'
            return (-f(x, z+2*h) + 4*f(x, z+h) - 3*f(x, z)) / (2*h)
    
    def backward_2nd(f, x, h, coord='x'):
        """2nd order backward difference"""
        if coord == 'x':
            return (3*f(x, z) - 4*f(x-h, z) + f(x-2*h, z)) / (2*h)
        else:  # coord == 'z'
            return (3*f(x, z) - 4*f(x, z-h) + f(x, z-2*h)) / (2*h)
    
    # Compute dc/dx based on position
    x_min, x_max = x_bounds
    if x + 2*h > x_max:  # Near right boundary
        dcdx = backward_2nd(velocity_2D, x, h, 'x')
    elif x - 2*h < x_min:  # Near left boundary
        dcdx = forward_2nd(velocity_2D, x, h, 'x')
    else:  # Interior points
        dcdx = central_4th(velocity_2D, x, h, 'x')
    
    # Compute dc/dz based on position
    z_min, z_max = z_bounds
    if z + 2*h > z_max:  # Near bottom boundary
        dcdz = backward_2nd(velocity_2D, z, h, 'z')
    elif z - 2*h < z_min:  # Near top boundary
        dcdz = forward_2nd(velocity_2D, z, h, 'z')
    else:  # Interior points
        dcdz = central_4th(velocity_2D, z, h, 'z')
    
    return np.array([dcdx, dcdz])

# --- Ray equations (Hamiltonian system) ---
def ray_equations(s, y):
    """y = [x, z, px, pz, T]."""
    x, z, px, pz, T = y
    c = velocity_2D(x, z)
    grad = grad_c(x, z)

    # Normalize slowness vector
    p_mag = np.sqrt(px**2 + pz**2)
    # Ray equations
    dxds = c**2 * px
    dzds = c**2 * pz
    dpxds = - (px**2 + pz**2) * c * grad[0]
    dpzds = - (px**2 + pz**2) * c * grad[1]
    dTds  = c * p_mag  # dT/ds = 1/(v * cosθ)
    return [dxds, dzds, dpxds, dpzds, dTds]

# --- Initialize ray ---
def shoot_ray(theta0_deg, x0=0.0, z0=0.0, s_max=10.0, npts=500):
    theta = np.radians(theta0_deg)
    c0 = velocity_2D(x0, z0)
    px0 = np.sin(theta) / c0
    pz0 = np.cos(theta) / c0
    y0 = [x0, z0, px0, pz0, 0.0]
    s_span = (0.0, s_max)
    s_eval = np.linspace(0.0, s_max, npts)
    sol = solve_ivp(ray_equations, s_span, y0, t_eval=s_eval, rtol=1e-8, atol=1e-10)
    return sol
```

```python

# --- Plot velocity model and ray paths ---
# Create grids for plotting
x = np.linspace(-10, 10, 100)
z = np.linspace(0, 10, 100)
X, Z = np.meshgrid(x, z)
V = velocity_2D(X, Z)

# Plot velocity model and ray paths
plt.figure(figsize=(15, 6))

# Plot velocity field
plt.subplot(121)
plt.pcolor(X, Z, V, shading='auto', cmap='seismic_r')  # seismic_r shows low velocities in red, high in blue
plt.colorbar(label='Velocity (km/s)')
plt.title('Velocity Model\n$c(x,z) = c_0(1 + 0.1\sin(0.5x) + 0.2z)$')
plt.xlabel('x [km]')
plt.ylabel('z [km]')
plt.gca().invert_yaxis()

# Plot ray paths
plt.subplot(122)
angles = np.linspace(-30, 30, 10)
for ang in angles:
    sol = shoot_ray(ang, x0=-2.5, z0=0)
    x, z, T = sol.y[0], sol.y[1], sol.y[4]
    plt.plot(x, z, label=f"{ang:.0f}° (T={T[-1]:.2f}s)")

plt.gca().invert_yaxis()
plt.xlabel('x [km]')
plt.ylabel('z [km]')
plt.title('Ray Paths')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

## 4.2 Create a 2D Horizontal Velocity Model

We'll create a velocity model in the horizontal plane with realistic velocity anomalies representing geological structures.

```python
# Define the model domain
nx, ny, nz = 201, 201, 2  # Grid dimensions (add thin z dimension)
x_min, x_max = -10.0, 10.0  # km
y_min, y_max = -10.0, 10.0  # km
z_min, z_max = 0.0, 1.0  # km (thin layer with constant properties)

# Create coordinate arrays
x = np.linspace(x_min, x_max, nx)
y = np.linspace(y_min, y_max, ny)
z = np.linspace(z_min, z_max, nz)
X, Y = np.meshgrid(x, y)

# Create 2D velocity model with Gaussian anomalies
c0 = 3.0  # Background velocity (km/s)
velocity_2d = c0 * np.ones((ny, nx))

# Add low velocity anomaly (sedimentary basin)
velocity_2d -= 0.5 * np.exp(-((X - 4.0)**2 + (Y - 2.0)**2) / 8.0)

# Add high velocity anomaly (basement rock)
velocity_2d += 0.6 * np.exp(-((X + 3.0)**2 + (Y + 3.0)**2) / 6.0)

# Add another low velocity anomaly
velocity_2d -= 0.4 * np.exp(-((X - 1.0)**2 + (Y - 6.0)**2) / 10.0)

# Add smooth gradient
velocity_2d += 0.2 * Y

# Create 3D velocity model by replicating the 2D model along z
# Shape will be (nx, ny, nz) for PyKonal
velocity_model = np.zeros((nx, ny, nz))
for k in range(nz):
    velocity_model[:, :, k] = velocity_2d.T  # Transpose to match PyKonal's (x, y, z) indexing

# Plot the velocity model (show z=0 slice)
plt.figure(figsize=(10, 8))
im = plt.pcolormesh(X, Y, velocity_2d, shading='auto', cmap='seismic_r')
plt.colorbar(im, label='Velocity (km/s)')
plt.xlabel('X (km)')
plt.ylabel('Y (km)')
plt.title('3D Horizontal Velocity Model (z=0 slice)\nConstant properties in depth')
plt.axis('equal')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"Velocity range: {velocity_model.min():.2f} - {velocity_model.max():.2f} km/s")
print(f"Grid dimensions: {nx} x {ny} x {nz}")
print(f"Grid spacing: dx={((x_max-x_min)/(nx-1)):.3f} km, dy={((y_max-y_min)/(ny-1)):.3f} km, dz={((z_max-z_min)/(nz-1)):.3f} km")
```

## 4.3 Set Up PyKonal Solver

PyKonal uses the Fast Marching Method to solve the eikonal equation $|\nabla T|^2 = 1/c^2(\mathbf{x})$ on a regular grid. We need to:
1. Create a PyKonal solver object
2. Set the velocity model
3. Define the source location

```python
# Create PyKonal solver for 3D problem
solver = pykonal.EikonalSolver(coord_sys="cartesian")

# Define the grid
solver.velocity.min_coords = x_min, y_min, z_min
solver.velocity.node_intervals = (x_max - x_min) / (nx - 1), (y_max - y_min) / (ny - 1), (z_max - z_min) / (nz - 1)
solver.velocity.npts = nx, ny, nz

# Set the velocity model (PyKonal expects velocity, not slowness)
solver.velocity.values = velocity_model

# Set step size for ray tracing (controls how finely rays are sampled)
# This is a property of the velocity field
# solver.velocity.step_size = 0.05  # 50 meters step size for smooth ray paths

# Define source location (center of domain at z=0) - USE PHYSICAL COORDINATES
source_x, source_y, source_z = 0, 0, 0
# find indexes corresponding to physical coordinates
idx_x = int((source_x - x_min) / (x_max - x_min) * (nx - 1))
idx_y = int((source_y - y_min) / (y_max - y_min) * (ny - 1))
idx_z = int((source_z - z_min) / (z_max - z_min) * (nz - 1))
src_idx = (idx_x, idx_y, idx_z)
solver.traveltime.values[src_idx] = 0.0  # Set travel time at source to zero

# Set the unknown flag for the source node to False.
# This is an FMM state variable indicating which values are completely
# unknown. Setting it to False indicates that the node has a tentative value
# assigned to it. In this case, the tentative value happens to be the true,
# final value.
solver.unknown[src_idx] = False
# Push the index of the source node onto the narrow-band heap.
solver.trial.push(*src_idx)


print(f"Source location (physical coords): ({source_x:.1f}, {source_y:.1f}, {source_z:.1f}) km")
print(f"Velocity at source: {velocity_model[nx//2, ny//2, 0]:.2f} km/s")
print(f"Ray tracing step size: {solver.velocity.step_size:.3f} km")
print("PyKonal solver configured successfully!")
```

## 4.8 Summary and Key Observations

This exercise demonstrated:

1. **PyKonal Eikonal Solver**: 
   - Fast Marching Method for computing travel times on regular grids
   - More efficient than ray tracing for dense sensor arrays
   - Automatically finds first-arrival travel times

2. **Velocity Model Effects**:
   - Low velocity anomalies increase travel times (positive residuals)
   - High velocity anomalies decrease travel times (negative residuals)
   - Ray paths bend around velocity heterogeneities (Fermat's principle)

3. **Applications**:
   - **Seismic Tomography**: Use travel time residuals to invert for velocity structure
   - **Earthquake Location**: Given travel times at sensors, locate unknown source
   - **Microseismic Monitoring**: Real-time source location for induced seismicity

4. **Advantages of Eikonal Solvers**:
   - Compute travel times to all points simultaneously
   - No need for iterative ray shooting
   - Stable for complex velocity models
   - Efficient for forward modeling in tomography
