# 1. Ray Tracing 

## 1.1 Scalar wave equation

$$
\nabla^2 u - \frac{1}{c^2(\mathbf{x})}\frac{\partial^2 u}{\partial t^2}=0.
$$
We assume a plane wave  $u=A(\mathbf{x})e^{i\omega(T(\mathbf{x})-t)}$.
Plugging in:
$$
\nabla^2(Ae^{i\omega(T-t)}) + \frac{\omega^2}{c^2}Ae^{i\omega(T-t)}=0.
$$

Expanding:
$$
\nabla^2A + 2i\omega \left[ \nabla A\cdot\nabla T + A\nabla^2T \right] - \omega^2 \left[ A|\nabla T|^2 + \frac{1}{c^2}A \right] = 0.
$$

## 1.2 Separate by powers of frequency: The WKB Approximation

The WKB (Wentzel-Kramers-Brillouin) approximation is a method for finding approximate solutions to linear differential equations with spatially varying coefficients. In seismology, it's particularly useful because:

1. **Frequency Dependence**: It assumes that the wavelength is much shorter than the scale over which the medium properties (like velocity) vary: $\lambda \ll L$ where $L$ is the characteristic length of velocity variation.

2. **Physical Meaning**: This is equivalent to assuming:
   - The wave behaves locally like a plane wave
   - Wave properties (amplitude, phase) vary slowly compared to the wavelength
   - Frequency $\omega$ is large compared to spatial variations

3. **Mathematical Procedure**: We separate the solution into:
   - A rapidly varying phase term $e^{i\omega T(\mathbf{x})}$
   - A slowly varying amplitude term $A(\mathbf{x})$

This leads to our expansion in powers of frequency $\omega$. 

### (a) High frequency Approximation ($\omega^2$ terms)

The Eikonal equation emerges from the highest order terms ($\omega^2$):

$$
|\nabla T|^2 = \frac{1}{c^2(\mathbf{x})}.
$$

This defines:
- Travel time $T(\mathbf{x})$ along ray paths
- Slowness vector $\mathbf{p}=\nabla T$ (direction of wave propagation)
- Local wave speed $c(\mathbf{x})$

The Eikonal equation is a nonlinear first-order PDE that:
1. Determines the phase of the wavefield
2. Defines ray paths as characteristics
3. Represents geometric optics limit of wave propagation

### (b) Order ($\omega$): Transport equation

The transport equations will constrain the amplitude decay as a function of distance.

$$
2\nabla A\cdot\nabla T + A\nabla^2 T = 0.
$$
Along a ray path where $\frac{d\mathbf{x}}{ds} = \frac{\nabla T}{|\nabla T|}$:
$$
\frac{dA}{ds} = -\frac{A}{2}\nabla^2 T , \frac{1}{|\nabla T|}.
$$
Integrating gives $A^2J = \text{const}$, with $J$ the geometric spreading (Jacobian of nearby rays).

## 1.3 Ray equations as a Hamiltonian system

The Hamiltonian formulation provides a powerful framework for understanding seismic wave propagation and ray tracing. Here's a detailed breakdown:

### What is a Hamiltonian?

The Hamiltonian $H$ is a function that describes the total energy of a system in terms of positions and momenta. In seismic wave propagation:

$$
H(\mathbf{x}, \mathbf{p}) = \frac{1}{2}\sum_i p_i^2 - \frac{1}{2}c^2(\mathbf{x})|\mathbf{p}|^2
$$

where:
- $\mathbf{x}$ are the spatial coordinates
- $\mathbf{p}$ are the components of slowness vector (momentum)
- $c(\mathbf{x})$ is the velocity field

### Hamilton's Equations

Define Hamiltonian $H(\mathbf{x},\mathbf{p})=\frac12(c^2|\mathbf{p}|^2-1)$.
Then the rate of change in position and wavenumber is:
$$
\frac{d\mathbf{x}}{dt} = \frac{\partial H}{\partial \mathbf{p}} = c^2\mathbf{p}, \quad
\frac{d\mathbf{p}}{dt} = -\frac{\partial H}{\partial \mathbf{x}} = -(\nabla c)c|\mathbf{p}|^2.
$$


The Hamiltonian formulation is equivalent to the Eikonal equation: The slowness vector $\mathbf{p} = \nabla T$ and the Hamiltonian constraint $H=0$ for physical ray paths (e.g., $\mathbf{x}$ is along ray paths).

With $(|\mathbf{p}|=1/c)$, we get the **ray curvature** equation:
$$
\frac{d\hat{\mathbf{n}}}{ds} = \nabla\ln c - (\hat{\mathbf{n}}\cdot\nabla\ln c)\hat{\mathbf{n}}.
$$


The ray paths are determined by Hamilton's canonical equations:

$$
\begin{align}
\dot{\mathbf{x}} &= \frac{\partial H}{\partial \mathbf{p}} \\
\dot{\mathbf{p}} &= -\frac{\partial H}{\partial \mathbf{x}}
\end{align}
$$

These equations give us:
1. The rate of change of position ($\dot{\mathbf{x}}$) is the group velocity
2. The rate of change of slowness ($\dot{\mathbf{p}}$) depends on velocity gradients


## 1.4 Analytical examples

### (a) Homogeneous medium 

Assume $c=c_0$.

Eikonal: $(|\nabla T|=1/c_0)$.
Solution: $(T(\mathbf{x}) = |\mathbf{x}-\mathbf{x}_s|/c_0)$.
Rays: straight lines.

Amplitude:
$$
A \propto
\begin{cases}
1/\sqrt{r} & 2\text{-D}\
1/r & 3\text{-D.}
\end{cases}
$$

### (b) Linear velocity gradient 

Assume $(c(z)=c_0+\gamma z)$.

Horizontal slowness $p$ conserved: $(p=\sin\theta/c(z))$.
Then $ \cos\theta(z)=\sqrt{1-p^2c^2(z)}$ and we integrate:

$$
x(z) = \int \frac{pc(z)}{\sqrt{1-p^2c^2(z)}}dz,
\qquad
T(z) = \int \frac{dz}{c(z)\sqrt{1-p^2c^2(z)}}.
$$

Substitute $(c(z)=c_0+\gamma z$)$ ⇒ logarithmic form:
$$
x(z)=\frac{1}{\gamma p}\sqrt{1-p^2c^2(z)}-\frac{1}{\gamma p}\sqrt{1-p^2c_0^2}.
$$
Travel time follows similarly (elliptic or log term). Rays are circular arcs.


### (c) Layered medium 

Assume $c=c(z)$, with piece-wise constant velocities.

Snell’s law: $p=\sin\theta/c(z)=\text{const}$.

Between two interfaces $(z_1,z_2)$:
$$
\Delta x = \int_{z_1}^{z_2} \frac{pc(z)}{\sqrt{1-p^2c^2(z)}},dz,
\qquad
\Delta T = \int_{z_1}^{z_2} \frac{1}{c(z)\sqrt{1-p^2c^2(z)}},dz.
$$

### Integration for layered medium with surface source-receiver geometry

For a source and receiver both at the surface $(z=0)$, with horizontal offset $X$, we need to:

1. **Determine maximum depth**: For a given source-receiver distance $X$, the ray parameter $p$ must satisfy:
   $$
   p \leq \frac{1}{v_{\text{max}}}
   $$
   where $v_{\text{max}}$ is the maximum velocity encountered along the ray path.

2. **Ray path equations**: For each layer $i$ with velocity $c_i$ and thickness $h_i$:
   $$
   \Delta x_i = \frac{2ph_i}{\sqrt{1-p^2c_i^2}}
   $$
   $$
   \Delta T_i = \frac{2h_i}{c_i^2\sqrt{1-p^2c_i^2}}
   $$
   where the factor of 2 accounts for down-going and up-going paths.

3. **Finding the correct $p$**: Solve for $p$ that satisfies:
   $$
   X = \sum_{i=1}^n \Delta x_i(p)
   $$
   This is typically done numerically since the equation is nonlinear.

4. **Turning depth**: For a given $p$, the turning depth occurs in the layer where:
   $$
   c(z_{\text{turn}}) = \frac{1}{p}
   $$

### Example for two-layer case:
For a model with:
- Layer 1: $0 \leq z \leq h_1$, velocity $c_1$
- Layer 2: $h_1 \leq z \leq h_2$, velocity $c_2 > c_1$

The total horizontal distance $X$ is:
$$
X = \begin{cases}
\frac{2ph_1}{\sqrt{1-p^2c_1^2}} & \text{if ray turns in layer 1} \\
\frac{2ph_1}{\sqrt{1-p^2c_1^2}} + \frac{2ph_2}{\sqrt{1-p^2c_2^2}} & \text{if ray turns in layer 2}
\end{cases}
$$

The corresponding travel time $T$ is:
$$
T = \begin{cases}
\frac{2h_1}{c_1^2\sqrt{1-p^2c_1^2}} & \text{if ray turns in layer 1} \\
\frac{2h_1}{c_1^2\sqrt{1-p^2c_1^2}} + \frac{2h_2}{c_2^2\sqrt{1-p^2c_2^2}} & \text{if ray turns in layer 2}
\end{cases}
$$

````python
import numpy as np
from scipy.optimize import root_scalar
import matplotlib.pyplot as plt

def compute_X_T(p, layers):
    """
    Compute total horizontal distance X and travel time T for a given ray parameter
    
    Parameters:
    -----------
    p : float
        Ray parameter (horizontal slowness)
    layers : list of tuples
        Each tuple contains (thickness, velocity) for each layer
    
    Returns:
    --------
    X : float
        Total horizontal distance
    T : float
        Total travel time
    """
    X = 0
    T = 0
    
    for h, c in layers:
        # Skip layer if ray parameter exceeds critical value
        if p > 1/c:
            return np.inf, np.inf
        
        # Add contribution from this layer
        X += 2 * p * h / np.sqrt(1 - p**2 * c**2)
        T += 2 * h / (c**2 * np.sqrt(1 - p**2 * c**2))
    
    return X, T

# Define a two-layer model
layers = [
    (2.0, 3.0),  # 2 km thick, 3 km/s
    (3.0, 5.0)   # 3 km thick, 5 km/s
]

# Range of source-receiver distances to consider
X_range = np.linspace(0.1, 20, 50)
T_values = []

for X_target in X_range:
    # Define objective function for root finding
    def objective(p):
        X, _ = compute_X_T(p, layers)
        return X - X_target
    
    # Find ray parameter that gives desired X
    try:
        # Bracket the solution between minimum and maximum possible p
        p_max = 1/min(v for h, v in layers)
        result = root_scalar(objective, bracket=[1e-6, p_max])
        p = result.root
        
        # Compute travel time for this ray parameter
        _, T = compute_X_T(p, layers)
        T_values.append(T)
    except:
        T_values.append(np.nan)

# Plot travel time curve
plt.figure(figsize=(10, 6))
plt.plot(X_range, T_values, 'b-', label='Travel Time Curve')
plt.xlabel('Source-Receiver Distance (km)')
plt.ylabel('Travel Time (s)')
plt.title('Travel Time Curve for Two-Layer Model')
plt.grid(True)
plt.legend()
plt.show()
````

### Understanding the Implementation

The code above demonstrates:

1. **Ray Parameter Search**: For each source-receiver distance $X$, we find the ray parameter $p$ that satisfies the total distance equation using root-finding.

2. **Layer Integration**: For each layer, we compute:
   - The horizontal distance contribution $\Delta x_i$
   - The travel time contribution $\Delta T_i$
   
3. **Critical Angles**: The code handles critical ray parameters where $p = 1/c_i$ for any layer $i$. Beyond these values, the equations become complex.

4. **Travel Time Curve Features**:
   - Near-offset (small $X$): Nearly vertical rays, $T \approx 2z/c_1$
   - Far-offset: Rays turning in deeper layers, showing characteristic curve shape
   - Discontinuities in $dT/dX$ indicate critical angles

This implementation can be extended to:
- Add more layers
- Include velocity gradients within layers
- Calculate amplitude effects
- Handle diving waves and head waves

## 5. Amplitude from geometrical spreading

Let $J(s)$ be the cross-sectional area of a ray tube.

Energy conservation constrains that  $A^2 v_g J = \text{const}$, with $v_g=c$.
$$
A(s) = A_0\left(\frac{J_0c_0}{J(s)c(s)}\right)^{1/2}.
$$

In 2-D, $J\propto \partial x_\perp/\partial\theta_0$, often computed numerically via paraxial rays.


# 2. Python Implementation: 2-D Ray Tracer

Below is a minimal script that integrates ray equations for a 2-D isotropic medium $c(x,z)$.



```python
#!/usr/bin/env python3
"""
2D Ray Tracing using the WKB/Hamiltonian equations
-------------------------------------------------
Solves for x(z) and travel time T in an arbitrary velocity field c(x,z).
Example velocity model: linear vertical gradient.
"""

import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# --- velocity model and its gradient ---
def velocity(x, z):
    """Velocity model c(x,z) [km/s]."""
    c0, grad = 3.0, 0.5  # km/s, km/s per km
    return c0 + grad * z

def grad_c(x, z):
    """Gradient of velocity [dc/dx, dc/dz]."""
    return np.array([0.0, 0.5])

# --- Ray equations (Hamiltonian system) ---
def ray_equations(s, y):
    """y = [x, z, px, pz, T]."""
    x, z, px, pz, T = y
    c = velocity(x, z)
    grad = grad_c(x, z)

    # Normalize slowness vector
    p_mag = np.sqrt(px**2 + pz**2)
    # Ray equations
    dxds = c**2 * px
    dzds = c**2 * pz
    dpxds = - (px**2 + pz**2) * c * grad[0]
    dpzds = - (px**2 + pz**2) * c * grad[1]
    dTds  = c * p_mag  # dT/ds = 1/(v * cosθ)
    return [dxds, dzds, dpxds, dpzds, dTds]

# --- Initialize ray ---
def shoot_ray(theta0_deg, x0=0.0, z0=0.0, s_max=20.0, npts=500):
    theta = np.radians(theta0_deg)
    c0 = velocity(x0, z0)
    px0 = np.sin(theta) / c0
    pz0 = np.cos(theta) / c0
    y0 = [x0, z0, px0, pz0, 0.0]
    s_span = (0.0, s_max)
    s_eval = np.linspace(0.0, s_max, npts)
    sol = solve_ivp(ray_equations, s_span, y0, t_eval=s_eval, rtol=1e-8, atol=1e-10)
    return sol

# --- Main: trace a family of rays ---
if __name__ == "__main__":
    angles = np.linspace(-30, 30, 7)  # degrees
    plt.figure(figsize=(7, 4))
    for ang in angles:
        sol = shoot_ray(ang, z0=0)
        x, z, T = sol.y[0], sol.y[1], sol.y[4]
        plt.plot(x, z, label=f"{ang:.0f}° (T_end={T[-1]:.2f}s)")
    plt.gca().invert_yaxis()
    plt.xlabel("x [km]")
    plt.ylabel("z [km]")
    plt.title("2D Ray Tracing (Linear Gradient Model)")
    plt.legend()
    plt.grid()
    plt.show()
```

### 🧠 Notes

* The ODEs evolve ((x,z,p_x,p_z,T)) along arclength (s).
* The code integrates up to `s_max`; you can terminate when (z) reaches a receiver depth.
* Replace `velocity()` and `grad_c()` with any 2-D grid or interpolation model.
* To add **reflection** at interfaces, invert (p_z) when (z=z_{\text{interface}}).
